```python
import pickle
import pandas as pd
import numpy as np
```

```python
mdl = pickle.load( open( "../model/mdl.obj", "rb" ) )
```

```python
df = pd.read_parquet(path='../data/predict.parquet', engine='pyarrow')
```

```python
df.head()
```

```python
features = np.delete(arr=df.columns.values, obj=np.where(df.columns.values == 'turnover'))
```

```python
features
```

```python
print(mdl)

# make predictions
pred = mdl.predict(df[features])
```

```python
pd.DataFrame(
    data={'turnover': pred}
).to_csv('results.csv')
```
